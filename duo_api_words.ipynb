```python
import pandas as pd
import numpy as np

import os
from dotenv import load_dotenv
load_dotenv()
username = os.environ.get("USERNAME")
password = os.environ.get("PASSWORD")

import duolingo as duo
```

```python
lingo  = duo.Duolingo(username, password)
words_json = lingo.get_vocabulary('de') #returns https://www.duolingo.com/vocabulary/overview json
```

```python
vocab_normal = pd.json_normalize(words_json,record_path=['vocab_overview'])
vocab_normal.head(10)
```

```python
vocab_normal.drop(columns=['strength_bars','last_practiced_ms','last_practiced','strength','id']).to_csv('duolingo_words.csv',index=False)
```
